```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
```

```python
start =datetime.datetime(2021,2,1)
end =datetime.datetime(2021,2,28)
```

```python
google=web.DataReader("GOOGL","yahoo",start,end)
```

```python
google.head()
```

```python
google.tail()
```

```python
google["Open"].plot(label="Openning price of the google stock", figsize=(15,7))
google["Close"].plot(label="Closing price of the google stock")
google["Low"].plot(label="Lowest price of the google stock")
google["High"].plot(label="Highest price of the google stock")
plt.title("Google Stock Price Analysis")
plt.ylabel("Stock price in dollars")
plt.legend()
plt.show()
```

```python
google["Volume"].plot(label="Volume of stock traded", figsize=(15,7))
plt.legend()
plt.title("Volume of Google Stock Traded")
plt.ylabel("Volume traded")
plt.show()
```
